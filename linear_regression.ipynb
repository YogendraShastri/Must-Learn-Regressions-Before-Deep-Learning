```python
from sklearn.linear_model import LinearRegression
import numpy as np

house_sizes = np.array([[500],[750],[1000],[1200],[1500],[1750], [2000]])
house_prices = np.array([1500000, 2000000, 2500000, 2557000, 2750000, 3220000, 3500000])

model = LinearRegression()
model.fit(house_sizes, house_prices)
```

```python
# predict the price
new_size = np.array([[1600]])
predicted_price = model.predict(new_size)
print(f"Predicted Price : {predicted_price[0]:.2f} Rs")
```

```python
# plotting
import matplotlib.pyplot as plt

plt.scatter(house_sizes, house_prices, label='Training Data')
plt.scatter(new_size, predicted_price, label="predicted data", color='red')
plt.plot(house_sizes, model.predict(house_sizes), label="RegressionLine", color="green")
plt.xlabel("House Size (sq ft)")
plt.ylabel("House Prices")
plt.title("House price prediction")
plt.legend()
plt.grid(True)
plt.show()
```
